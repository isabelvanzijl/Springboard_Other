# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

# B. Provide support for your arguments and your own recommendations with a statistically sound analysis

## Hypothesis test

$H_0$: There is no significant correlation between hospital capacity (number of discharges) and readmission rates

$H_A$: There is a significant correlation between hospital capacity (number of discharges) and readmission rates

## Compute and report the observed significance value (or p-value)

Let's try to do this with a permutation test.

```python
# extract relevant columns of data
num_discharges = clean_hospital_read_df['Number of Discharges']
excess_readmission_ratio = clean_hospital_read_df['Excess Readmission Ratio']

# calculate and print Pearson correlation coefficient
pearson_r = np.corrcoef(num_discharges, excess_readmission_ratio)[0, 1]
print('Pearson correlation coefficient: ', pearson_r)
```

Our Pearson correlation coefficient is returning a null value. This must be because some of our rows contain null values. We will have to clean the dataframe before continuing our analysis.

```python
# drop null rows from dataframe
clean_hospital_dropnull = clean_hospital_read_df[np.isfinite(clean_hospital_read_df['Excess Readmission Ratio'])]
clean_hospital_dropnull.head()
```

```python
# (repeat) extract relevant columns of data
num_discharges = clean_hospital_dropnull['Number of Discharges']
excess_readmission_ratio = clean_hospital_dropnull['Excess Readmission Ratio']

# (repeat) calculate and print Pearson correlation coefficient
pearson_r = np.corrcoef(num_discharges, excess_readmission_ratio)[0, 1]
print('Pearson correlation coefficient: ', round(pearson_r,3))
```

```python
# initiate empty permutation replicates array
size = 100000
perm_reps = np.empty(size)

# fill the permutation replicates array
for i in range(size):
    num_discharges_perm = np.random.permutation(num_discharges)
    perm_reps[i] = np.corrcoef(num_discharges_perm, excess_readmission_ratio)[0, 1]

# compute and print the p value
p = np.sum(perm_reps <= pearson_r) / len(perm_reps) 
print('p value:', round(p,3))
print('p value (unrounded):', p)
```

We get a very small p value (essentially zero), signaling that we should reject the null hypothesis. There is indeed a significant correlation between hospital capacity (number of discharges) and readmission rates.

## Report statistical significance for $\alpha$ = .01. 

Since the p value we calculated was essentially zero, we can conclude that the correlation is still statistically significant for $\alpha$ = .01.

## Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?

In terms of the Pearson correlation coefficient, statistical significance establishes the strength of the association between two continuous variables. Essentially, we want to find out if the association is due to outliers or if there's an actual trend between the two variables. We found a Pearson correlation coefficient of approximately -0.097, signaling that our two variables have a negative linear relationship. Our tiny p value reveals that this correlation between our two variables is statistically significant.

Practical significance is the determination of whether the correlation between two variables is actually important. Pearson correlation coefficients fall between the interval of [-1, +1], where 0 represents no presence of a correlation. Our coefficient value of approximately -0.097 is very close to zero, signaling that the correlation of these two variables is very slight/small. Even though our p value reveals strong statistical significance, the actual correlation trend is so small that the linear relationship between the number of discharges and readmission rates is practically insignificant.

To conclude, the number of discharges is not a good predictor for readmission rates. The conclusions and recommendations of the preliminary report shouldn't be followed. Further testing should be completed to determine which variables could be strong predictors for readmission rates.

A key visual that's missing from the plot above is a linear regression trend line. The shaded rectangles don't tell us anything, but a trend line would show us the actual relationship between the two variables.

```python
# build initial scatter plot
fig, ax = plt.subplots(figsize=(16,8))
plt.scatter(num_discharges, excess_readmission_ratio, alpha=0.2)

# add plot limits, labels, and title
ax.set_xlim([0, max(num_discharges)+500])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

# calculate linear regression fit line
a, b = np.polyfit(num_discharges, excess_readmission_ratio, 1)
x = np.array([0, 7000])
y = a * x + b

# plot linear regression fit line
plt.plot(x, y, color='red', alpha=.5)
plt.show()

# display linear fit line
print('Regression Line Equation: y =', round(a,5), 'x + ', round(b,5))
```

We can now clearly see the negative relationship between the number of discharges and the excess rate of readmissions.
